```python
import os
import pandas as pd
from DFCreation import extract_peaks
```

```python
# Initialize results list and counters
csvFileResults = []
invalid_files_list = []

valid_files = 0
invalid_files = 0

# Data directory location
dataLocation = "./Data/After_May"

totalFiles = len(os.listdir(dataLocation))

# Loop through files in the directory
for i, file in enumerate(os.listdir(dataLocation), start=1):
    # Get the full file path
    fullFilePath = os.path.join(dataLocation, file)
    
    # Call extract_peaks and handle the result
    fileResult = extract_peaks(fullFilePath)
    
    # Check if the result is a valid dictionary
    if isinstance(fileResult, dict):
        csvFileResults.append(fileResult)  # Append only valid results
        valid_files += 1  # Increment valid file counter
    else:
        print(f"Skipping file {file}: Invalid result returned.")
        invalid_files += 1  # Increment invalid file counter
        invalid_files_list.append({"file_name": file})  # Add invalid file to the list
    
    # Print progress
    totalFilesProcessed = valid_files + invalid_files
    print(f"Processed {totalFilesProcessed}/{totalFiles} files...")

# Calculate and print percentage of valid files
if totalFilesProcessed > 0:
    valid_percentage = (valid_files / totalFilesProcessed) * 100
    print(f"Valid files: {valid_files}/{totalFilesProcessed} ({valid_percentage:.2f}%)")
else:
    print("No files processed.")

# Save valid results to CSV
if csvFileResults:  # Ensure there is data to save
    ClusteringDF = pd.DataFrame(csvFileResults)
    ClusteringDF.to_csv('../Data/ClusteringDF.csv', index=False)
    print("ClusteringDF.csv saved successfully!")
else:
    print("No valid data to save. Check the logs.")

# Save invalid files to a separate CSV
if invalid_files_list:  # Ensure there are invalid files to save
    InvalidFilesDF = pd.DataFrame(invalid_files_list)
    InvalidFilesDF.to_csv('../InvalidFiles/InvalidFiles.csv', index=False)
    print("InvalidFiles.csv saved successfully!")
else:
    print("No invalid files to save.")
```
